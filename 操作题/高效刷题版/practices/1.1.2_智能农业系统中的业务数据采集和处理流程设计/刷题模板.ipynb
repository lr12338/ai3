## 数据集说明

- `SensorID`：传感器 ID
- `Timestamp`：时间戳
- `SensorType`：Temperature / Humidity / SoilMoisture / SoilPH / Light
- `Value`：读数
- `Location`：Field1–Field4


```python
import pandas as pd
import numpy as np

data = pd.read_csv('data/sensor_data.csv')
data.head()
```

## 任务1：统计每种传感器的数据数量和平均值

```python
# sensor_stats = data.groupby('SensorType')['Value'].agg(['count', 'mean'])
# print("传感器数据数量和平均值:")
# print(sensor_stats)

# stats = data.groupby("分组列")["指标列"].agg(["count", "mean"])
sensor_status = data.groupby('SensorType')['Value'].agg(['count', 'mean'])
# .groupby('SeneorType') 按传感器类型分组
# ['Value'] 指标列
# .agg(['count', 'mean']) 对每个小组做两种统计：计数 + 平均值

print(sensor_status)
```

## 任务2：统计每个位置的温度和湿度传感器平均值

```python
location_stats = (
    data[data['SensorType'].isin(['Temperature', 'Humidity'])] # data[筛选条件] data['a'].isin(['b','c'])只保留a列中 b、c的数据行    
    .groupby(['Location', 'SensorType'])['Value'] # .groupby(['a','b']) 按a、b列分组, ['c']只对 c这一列做后续计算
    .mean() #对['Value']列进行平均值计算
    .unstack() #“展开” 索引，把行索引变成列，透视
)
print("每个位置的温度和湿度数据平均值:")
print(location_stats)

```

## 任务3：异常标记与缺失值处理并导出

```python
data['is_abnormal'] = np.where(
    ((data['SensorType'] == 'Temperature')
     & ((data['Value'] < -10) | (data['Value'] > 50)))
    | (
        (data['SensorType'] == 'Humidity')
        & ((data['Value'] < 0) | (data['Value'] > 100))
    ),
    True,
    False,
)

print("异常值数量:", data["is_abnormal"].sum())

data['Value'] = data['Value'].ffill().bfill()

cleaned_data = data.drop(columns=['is_abnormal'])
cleaned_data.to_csv('cleaned_sensor_data.csv', index=False)
print("数据清洗完成，已保存为 cleaned_sensor_data.csv")
```
