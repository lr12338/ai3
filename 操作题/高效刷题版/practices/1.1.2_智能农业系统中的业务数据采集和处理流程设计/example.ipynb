# 1.1.2 智能农业系统中的业务数据采集和处理流程设计

## 任务说明
某农业公司计划引入智能农业系统，通过安装在农田中的多种传感器实时监控农田环境，收集温度、湿度、土壤水分、土壤酸碱度和光照等数据，为作物管理和产量优化提供支持。


## 数据集说明

- `SensorID`：传感器 ID
- `Timestamp`：时间戳
- `SensorType`：传感器类型（Temperature、Humidity、SoilMoisture、SoilPH、Light）
- `Value`：传感器读数
- `Location`：传感器安装位置（Field1-Field4）


```python
import pandas as pd
import numpy as np

# 读取数据集
data = pd.read_csv('data/sensor_data.csv')
data.head()
```

## 任务1：统计每种传感器的数据数量和平均值


```python
# 对传感器类型分组，统计数量和平均值
sensor_stats = data.groupby('SensorType')['Value'].agg(['count', 'mean'])

print('传感器数据数量和平均值:')
print(sensor_stats)
```

## 任务2：统计每个位置的温度和湿度传感器平均值


```python
# 先筛选温度和湿度，再按位置和传感器类型分组统计均值
location_stats = (
    data[data['SensorType'].isin(['Temperature', 'Humidity'])]
    .groupby(['Location', 'SensorType'])['Value']
    .mean()
    .unstack()
)

print('每个位置的温度和湿度数据平均值:')
print(location_stats)
```

## 任务3：数据清洗和异常值处理


```python
# 标记明显异常的温度和湿度数据
data['is_abnormal'] = np.where(
    ((data['SensorType'] == 'Temperature') & ((data['Value'] < -10) | (data['Value'] > 50)))
    |
    ((data['SensorType'] == 'Humidity') & ((data['Value'] < 0) | (data['Value'] > 100))),
    True,
    False,
)

print('异常值数量:', data['is_abnormal'].sum())

# 用前向填充和后向填充处理缺失值
data['Value'] = data['Value'].fillna(method='ffill')
data['Value'] = data['Value'].fillna(method='bfill')

# 删除辅助列并保存清洗结果
cleaned_data = data.drop(columns=['is_abnormal'])
cleaned_data.to_csv('cleaned_sensor_data.csv', index=False)
print("数据清洗完成，已保存为 'cleaned_sensor_data.csv'")
```
