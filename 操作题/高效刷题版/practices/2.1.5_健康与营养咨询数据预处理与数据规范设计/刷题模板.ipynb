## 数据：`data/健康咨询客户数据集.csv`

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

data = pd.read_csv('data/健康咨询客户数据集.csv')
print(data.info())
print(data.isnull().sum())

data_cleaned = data.dropna()

data_cleaned.loc[:, 'Your age'] = pd.to_numeric(
    data_cleaned['Your age'], errors='coerce'
)
data_cleaned = data_cleaned.dropna(subset=['Your age'])
data_cleaned = data_cleaned[data_cleaned['Your age'] >= 0]
data_cleaned.loc[:, 'Your age'] = data_cleaned['Your age'].astype(int)

dup_cnt = data_cleaned.duplicated().sum()
data_cleaned = data_cleaned.drop_duplicates()
print(f"Removed {dup_cnt} duplicate rows")

fit_col = 'How do you describe your current level of fitness ?'
le = LabelEncoder()
data_cleaned[fit_col] = le.fit_transform(data_cleaned[fit_col].astype(str))

data.columns = data.columns.str.strip()
print(list(data.columns))

data_cleaned = data.dropna(subset=['How often do you exercise?'])
counts = data_cleaned['How often do you exercise?'].value_counts()
plt.figure(figsize=(10, 6))
counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Exercise Frequency')
plt.ylabel('')
plt.show()

data_filled = data.apply(lambda x: x.fillna(x.mode()[0]))
train_data, test_data = train_test_split(
    data_filled, test_size=0.2, random_state=42
)
out = '2.1.5_cleaned_split.csv'
pd.concat([train_data, test_data]).to_csv(out, index=False)
print("已保存", out)
```
