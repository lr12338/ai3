## 任务1：统计住院天数超过 7 天的患者数量及占比，并划分风险等级

```python
# data['RiskLevel'] = np.where(
#     data['DaysInHospital'] > 7,
#     '高风险患者',
#     '低风险患者',
# )

# risk_counts = data['RiskLevel'].value_counts()
# total = len(data)
# high_risk_ratio = risk_counts['高风险患者'] / total
# low_risk_ratio = risk_counts['低风险患者'] / total

# print("高风险患者数量:", risk_counts["高风险患者"])
# print("低风险患者数量:", risk_counts["低风险患者"])
# print("高风险患者占比:", high_risk_ratio)
# print("低风险患者占比:", low_risk_ratio)

# 任务1：统计住院天数超过 7 天的患者数量及占比，并划分风险等级
data['RiskLevel'] = np.where( data['DaysInHospital'] > 7, '高风险患者', '低风险患者' )
# np.where(data[''],true,false)
# data['新数据列'] = np.where( 条件， 正直， 负值）
riskcounts = data['RiskLevel'].value_counts() # 统计每个类别出现多少次  .value_counts()

total = len(data) # 统计总人数

highrisk = riskcounts['高风险患者'] / total # riskcounts['']

lowrisk = riskcounts['低风险患者'] / total

print(f"高风险患者数量: {riskcounts['高风险患者']}")
print(f"低风险患者数量: {riskcounts['低风险患者']}")
print(f"高风险患者占比: {highrisk:.2%}")
print(f"低风险患者占比: {lowrisk:.2%}")
```
