## 数据：`data/medical_data.csv`

```python
import pandas as pd
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

data = pd.read_csv('data/medical_data.csv', encoding='utf-8')
print(data.dtypes)
print(data.info())
print(data.isnull().sum())

data['就诊日期'] = pd.to_datetime(data['就诊日期'])
data['诊断日期'] = pd.to_datetime(data['诊断日期'])
data.rename(columns={'病人ID': '患者ID'}, inplace=True)
print(data.head())

data['诊断延迟'] = (data['诊断日期'] - data['就诊日期']).dt.days
data['病程'] = (datetime(2024, 9, 1) - data['诊断日期']).dt.days

data = data[
    (data['诊断延迟'] >= 0)
    & (data['病程'] > 0)
    & (data['年龄'] < 120)
]
print(data.describe())

initial_rows = data.shape[0]
data.drop_duplicates(inplace=True)
print(f"删除的重复行数: {initial_rows - data.shape[0]}")

scaler = MinMaxScaler()
cols_norm = ['年龄', '体重', '身高']
data[cols_norm] = scaler.fit_transform(data[cols_norm])
print(data.head())

dist = data.groupby('疾病类型')['治疗结果'].value_counts().unstack(fill_value=0)

font_path = 'C:/Windows/Fonts/simhei.ttf'
my_font = fm.FontProperties(fname=font_path)

dist.plot(kind='bar', stacked=True)
plt.title('不同疾病类型的治疗结果分布', fontproperties=my_font)
plt.xlabel('疾病类型', fontproperties=my_font)
plt.ylabel('数量', fontproperties=my_font)
plt.xticks(rotation=45, fontproperties=my_font)
plt.legend(prop=my_font)
plt.tight_layout()
plt.show()

sev_code = pd.Categorical(data['疾病严重程度']).codes
plt.scatter(data['年龄'], sev_code)
plt.title('年龄与疾病严重程度（编码）', fontproperties=my_font)
plt.xlabel('年龄', fontproperties=my_font)
plt.ylabel('严重程度编码', fontproperties=my_font)
plt.show()

data.to_csv('2.1.4_cleaned_data.csv', index=False)
print("已保存 2.1.4_cleaned_data.csv")
```
