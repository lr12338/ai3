## 数据分析与 docx 输出（考场素材）

- 请将 **`智能健康监测系统数据集.xlsx`** 放在本案例文件夹下。
- 报告：`3.1.4-1.docx`；优化方案：`3.1.4-2.docx`。

分析维度提示：
1. 用户活动周期：一天中健康指标变化
2. 健康指标偏好：常用与少用监测项
3. 系统响应与准确性：响应时间与误报风险


```python
import pandas as pd
from pathlib import Path

xlsx = Path('智能健康监测系统数据集.xlsx')
if not xlsx.exists():
    raise FileNotFoundError(
        f'请将数据集复制到: {{xlsx.resolve().parent}}'
    )

df = pd.read_excel(xlsx)
print("形状:", df.shape)
print(df.head())
print(df.describe(include='all'))

obj_cols = df.select_dtypes(include=['object']).columns.tolist()
num_cols = df.select_dtypes(include=['number']).columns.tolist()
lines = []
lines.append('# 分析报告（草稿，请按实际列名改写）')
for title, cols in [('分类字段概览', obj_cols), ('数值字段概览', num_cols)]:
    lines.append(f'## {title}')
    for c in cols[:12]:
        lines.append(f'### {c}')
        if c in obj_cols:
            vc = df[c].astype(str).value_counts().head(8)
            lines.append(vc.to_string())
        else:
            lines.append(df[c].describe().to_string())
report_body = '\n\n'.join(lines)
opt_body = (
    '# 优化方向（示例）\n\n'
    '1. 响应链路：缩短解析与设备回调耗时，加入超时重试与本地缓存。\n'
    '2. 个性化：基于高频场景做快捷入口与默认策略，减少操作步数。\n'
    '3. 稳定性：对异常值与弱网场景做降级策略，完善日志与远程诊断。\n'
)

try:
    from docx import Document
except ImportError:
    print('请安装: pip install python-docx openpyxl pandas')
else:
    def write_simple_doc(path: Path, text: str) -> None:
        doc = Document()
        for block in text.split('\n\n'):
            block = block.strip()
            if not block:
                continue
            if block.startswith('# '):
                doc.add_heading(block[2:].strip(), level=1)
            elif block.startswith('## '):
                doc.add_heading(block[3:].strip(), level=2)
            elif block.startswith('### '):
                doc.add_heading(block[4:].strip(), level=3)
            else:
                doc.add_paragraph(block)
        doc.save(path)

    write_simple_doc(Path('3.1.4-1.docx'), report_body)
    write_simple_doc(Path('3.1.4-2.docx'), opt_body)
    print("已生成:", '3.1.4-1.docx', '3.1.4-2.docx')
```
